## droped responses that has traits with less than 9 responses

```python
sns.countplot(responses['tech field'] )
```

```python
responses['tech field'].value_counts()
```

```python
tect_field_to_be_removed=[i for i in responses['tech field'].value_counts()[11:16].index]
```

```python
rows_to_be_removed=responses[responses['tech field'] == '']

for i in range(5):
    a=responses[responses['tech field'] == tect_field_to_be_removed[i]]
    rows_to_be_removed=pd.concat([rows_to_be_removed,a])
```

```python
rows_to_be_removed
```

```python
rows_to_be_removed.index
```

```python
responses=responses.drop(index=rows_to_be_removed.index)
```
